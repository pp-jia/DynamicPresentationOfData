# 动态条形图

## 动态条形图-改写 

### 随机生成数据 （改为自己从58同城下载的数据）

```python
import random
import pandas as pd
import math
import csv 

# 获取已经提前爬取好的数据
housePriceData = []
with open('./AllCityData.csv','r')as f:
    f_csv = csv.reader(f)
    next(f_csv)
    for row in f_csv:
        housePriceData.append(row[2])



cats0 = ['北京', '哈尔滨','上海','广州','深圳', '吉林', '沈阳','包头', '石家庄', '乌鲁木齐',
         '兰州', '西宁', '西安', '太原', '天津', '济南', '郑州', '成都', '拉萨', '重庆',
         '武汉', '合肥', '南京', '杭州', '南昌', '长沙', '贵阳', '昆明', '南宁', '福州', '海口']
color0 = ['#D6A5DD','#F7DC73','#EBBDBF','#76CBE8','#A4DFE5', '#7CFC00', '#7D9EC0', '#00FA9A', '#6959CD', '#FFD700',
          '#EE7AE9','#EE2C2C','#ADFF2F','#A020F0','#8B6508', '#EE00EE', '#EBEBEB', '#D2691E', '#CAFF70', '#BC8F8F',
          '#9A32CD', '#8B6508', '#8B008B', '#7CCD7C', '#424242', '#104E8B', '#00FFFF', '#C1C1C1', '#C67171', '#912CEE',
          '#8B0000']

year0 = pd.date_range(start='2015/1', end='2023/9', freq='MS').to_list()
year0_noday = []
for i in year0:
    date = i.date()
    year0_noday.append(date.strftime("%Y-%m"))

year = []
for i in year0_noday:
    for j in range(len(cats0)):
        year.append(i)

cats = cats0*len(year0_noday)
color = color0*len(year0_noday)

value = []

for i in range(len(year)):
#     print(housePriceData[int(i/5) + (i%5) * 105])   # 进行数据映射
    value.append(int(float(housePriceData[int(i/31) + (i%31) * 105])))#这里要把一些浮点数改为int，否则下边比较大小时有问题
    
df = pd.DataFrame()
df['year'] = year
df['value'] = value
df['cats'] = cats
df['color'] = color
# df['csum']=df.groupby(['cats'])['value']
```

### 把数据放到代码块中 

```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Timeline

tl = Timeline()
for i in year0_noday:
    df_sub = df[df['year']==i].sort_values(by='value')
    cats_list = list(df_sub['cats'])
    csum_list = list(df_sub['value'])
    color_list = list(df_sub['color'])

    bar = (
        Bar()
        .add_xaxis(cats_list)
        .add_yaxis("", csum_list,label_opts=opts.LabelOpts(position="right",font_weight="bold"),category_gap=30)
        .reversal_axis()
        .set_global_opts(
            title_opts=opts.TitleOpts("跑得快 (时间: {} 年)".format(i),pos_left=350,padding=[30,20]),
            xaxis_opts=opts.AxisOpts(max_=50000)
        )
    )
    tl.add(bar, "{}年".format(i))
    tl.add_schema(play_interval=300, is_loop_play=True)#播放间隔，是否循环播放
tl.render("timeline_bar_reversal_dynamic_data.html")
```

### 改变柱条颜色 

```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Timeline

tl = Timeline()
for i in year0_noday:
    df_sub = df[df['year']==i].sort_values(by='value')
    cats_list = list(df_sub['cats'])
    csum_list = list(df_sub['value'])
    color_list = list(df_sub['color'])
    y=[]
    for j in range(5):
        y.append(
            opts.BarItem(
                name=cats_list[j],
                value=csum_list[j],
                itemstyle_opts=opts.ItemStyleOpts(color=color_list[j]),
            )
        )

    bar = (
        Bar()
        .add_xaxis(cats_list)
        .add_yaxis("", y,label_opts=opts.LabelOpts(position="right",font_weight="bold"),category_gap=30)
        .reversal_axis()
        .set_global_opts(
            title_opts=opts.TitleOpts("跑得快 (时间: {} 年)".format(i),pos_left=350,padding=[30,20]),
            xaxis_opts=opts.AxisOpts(max_=50000000)
        )
    )
    tl.add(bar, "{}年".format(i))
    tl.add_schema(play_interval=300, is_loop_play=True)#播放间隔，是否循环播放
tl.render("timeline_bar_reversal_dynamic_color.html")
```

### 完整图 -添加图像后 

```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Timeline

tl = Timeline()
for i in year0_noday:
    df_sub = df[df['year']==i].sort_values(by='value')
    
    cats_list = list(df_sub['cats'])[16:]
    value_list = list(df_sub['value'])[16:]
    color_list = list(df_sub['color'])[16:]
    y=[]
    for j in range(15):
        y.append(
            opts.BarItem(
                name=cats_list[j],
                value=value_list[j],
                itemstyle_opts=opts.ItemStyleOpts(color=color_list[j]),
            )
        )
    data = []
    for k in range(15):
        data.append(opts.MarkPointItem(name="",coord=[value_list[k], cats_list[k]],symbol="null",symbol_size=0,
                                       itemstyle_opts=opts.ItemStyleOpts(color=color_list[k],border_color='grey')))
    bar = (
        Bar(
            init_opts=opts.InitOpts(width='95vw', height='2000px')
        )
        .add_xaxis(cats_list)
        .add_yaxis("", y,label_opts=opts.LabelOpts(position="right",font_weight="bold"),category_gap=3,
                  markpoint_opts=opts.MarkPointOpts(#每个条形图的分类坐标内容value=cats_list[0], value=cats_list[1], value=cats_list[2], value=cats_list[3], value=cats_list[4]
            data=data))
            #category_gap=30,
#         .add_yaxis("", csum_list, label_opts=opts.LabelOpts(position="right"))
        .reversal_axis()
        .set_global_opts(
            title_opts=opts.TitleOpts(title="房价数据 (时间: {})".format(i),pos_left=350,padding=[30,20]),
            legend_opts=opts.LegendOpts(is_show=True),
            xaxis_opts=opts.AxisOpts(max_=70000)
        )
    )
    tl.add(bar, "{}".format(i))
    tl.add_schema(play_interval=500, is_loop_play=False)#播放间隔，是否循环播放
tl.render("timeline_bar_reversal_dynamic_final.html")
```
